```python
import google.generativeai as genai

genai.configure(
    api_key="AQ.Ab8RN6Jn7AHuF-WiYKDXbdJD5EwbAa8xRF8dZHXJs942xC-2cg"
)

model = genai.GenerativeModel(
    "gemini-2.5-flash"
)
```

```python
reference_answer = """
Artificial Intelligence is a branch of computer science
that enables machines to simulate human intelligence.
"""

student_answer = """
AI allows machines to think and learn.
"""

prompt = f"""
You are an evaluator.

Reference Answer:
{reference_answer}

Student Answer:
{student_answer}

Generate:

1. Score out of 10
2. Strengths
3. Missing Concepts
4. Suggestions
"""

response = model.generate_content(
    prompt
)

print(response.text)
```
