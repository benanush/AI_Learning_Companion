```python
import os
import cv2
import matplotlib.pyplot as plt

DATASET_PATH = r"C:\Users\benan\OneDrive\Desktop\AI_Learning_Project\datastes\handwritten_ans"

images = [
    f for f in os.listdir(DATASET_PATH)
    if f.lower().endswith(".jpg")
]

print("Total Images:", len(images))

if len(images) > 0:
    sample = images[0]

    img = cv2.imread(os.path.join(DATASET_PATH, sample))

    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.figure(figsize=(8,6))
    plt.imshow(img)
    plt.title(sample)
    plt.axis("off")
    plt.show()
else:
    print("❌ No JPG files found inside handwritten_ans")
```
